As genres is given in string so to apply for over it we should have to convert it into list , so for that we use ast python library that converts from string to list

```python
import ast
import json
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

```python
def extract_names_from_json(input_string):
    try:
        # Parse the JSON string
        data = json.loads(input_string)
        
        # Extract the 'name' field from each dictionary
        names = [entry.get('name', '') for entry in data]
        
        # Return the list of names
        return names
    except json.JSONDecodeError:
        return []  # Return an empty list if there's an error during JSON decoding

```

```python
movies['cast']=movies['cast'].apply(extract_names_from_json)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def get_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
        
    return L
```

```python
movies['crew']=movies['crew'].apply(get_director)
```

```python
movies.head()
```

to make overview list from string

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.head()
```

now to remove spaces between names, that over tag our model should not get confuse if have same first/last names

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

## now we will convert tags from text to vector to calculate difference between them and can recommend

as there are some words having same meanings in features lie action , actions etc so we will remove them using nltk library(NLP)

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer
cv= CountVectorizer(max_features = 5000 , stop_words='english')
ps=PorterStemmer()
```

```python
def stem(text):
    L=[]
    for i in text.split(): # to convert string into list
        L.append(ps.stem(i))
        
    return " ".join(L) #to convert back list into string 
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
vectors = cv.fit_transform(df['tags']).toarray() #as countervecorization returns saipai  metrix so we will convert into numpy array using toarray() function
```

```python
def recomend(movie):
    movie_index=df[df['title'] == movie].index[0]
    distance = similarity[movie_index]
    recommended_movies = sorted(list(enumerate(distance)) , reverse=True , key= lambda x:x[1])[1:6]
    
    for i in recommended_movies:
        print(df.iloc[i[0]].title )
```
