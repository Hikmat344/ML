# now to perform preprocessing on the categorical data , to convert it into one hot encoding

```python
dataset1 = pd.get_dummies(df, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
```

```python
dataset1.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
standardscaler = StandardScaler()
```

```python
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[columns_to_scale] = standardscaler.fit_transform(dataset1[columns_to_scale])
dataset1.head()
```

```python
dataset1.rename(columns={'condition': 'target'} , inplace=True)
dataset1.head()
```

```python
y= dataset1['target']
x= dataset1.drop(['target'], axis=1)
```

```python
from sklearn.model_selection import cross_val_score
knn_scores=[]
for k in range(1,22):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score = cross_val_score(knn_classifier , x, y , cv=10)
    knn_scores.append(score.mean())
```

```python
plt.figure(figsize=(20,20))
plt.plot([k for k in range(1,22)] , knn_scores , color='green')
for i in range(1,22):
    plt.text(i , knn_scores[i-1] , ( i , knn_scores[i-1]))
plt.xticks([i for i in range(1,22)])
plt.xlabel('number of neighbors')
plt.ylabel('knn scores')
plt.suptitle('knn score for different k values')
```

```python
knn_classifier= KNeighborsClassifier(n_neighbors = 8)
score = cross_val_score(knn_classifier , x,y,cv=10)
score.mean()
```
