**Generator function** <br>
*iterate one by one
<br> *stop after each iteration <br>
*remember old (last) iteration value <br>
*next iterate go forward from that last one

```python
def my_range(start:int , end:int , step:int):
  for i in range(start , end+1 , step):
    yield i #generator function

a = list(my_range(1,10 , 1))
a
```

```python
from collections.abc import Iterator

def my_range(start:int , end:int , step:int = 1)->Iterator[int]:
  for i in range(start , end+1 , step):
    yield i #generator function

a :Iterator[int]= my_range(1,10 , 1)
print(next(a))
print(next(a))

print(type(a))

for i in a:
  print(i)
```

## Inheritance ( wirasat )

`class childclass(parentclass):` <br>

**Note:**
1. in inheritance if we make a method in child so this method will not be accessed in parent class
2. in parent class we write those features that are common in all stack holders of a class and which are individual cheristics that are write into that indivisual child class
3. if we want to inherit parent class in child class so then super() keyword is used for this purpose with __init__ (constructor ) of child class

```python
  class Parents():
    def __init__(self) -> None:
      self.eye_color : str = "brown"
      self.hair_color : str = "black"
    def speak(self , words: str) -> None:
      print(f"parent method speaks: {words}")
    def watch(self , object_name: str) -> None:
      print(f"you are looking {object_name}")
  class child(Parents):
    pass

  obj1: Parents = Parents()
  print(obj1.eye_color)
  print(obj1.hair_color)
  obj1.speak("pakistan")
  obj1.watch("tv")

  print("**************now child objects************")

# obj 2 child object , here there's no method in child itself but it is inherited from parent
  obj2: child = child()
  print(obj2.eye_color)
  print(obj2.hair_color)
  obj2.speak("Pakistan zindabad")
```
